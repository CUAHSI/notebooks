```python
import numpy as np
def Compute_AR_ID(ds, ar_con_var, new_var):

    ar_con_values = ds[ar_con_var].values
    ar_id = np.zeros_like(ar_con_values)

    # create a unique group ID for each consecutive group of one
    group_id = 0

    # we hit the first value and it is 1
    if ar_con_values[0] == 1:
            group_id = 1
            ar_dur[0] = group_id

    for i in range(1, len(ar_con_values)):
        # we hit the start of a new group
        if ar_con_values[i] == 1 and ar_con_values[i-1] == 0:
            group_id += 1
        # we hit a continuation of the previous group
        if ar_con_values[i] == 1:
            ar_id[i] = group_id
    
    # add the new variable to the dataset
    ds[new_var] = (['time'], ar_id)

    return ds
```

```python
def Compute_AR_ID(ds, ar_id_var='AR_ID', ivt_var='IVT', dur_var='AR_DUR', max_var='AR_MAX'):
    
    # get the AR_ID values from the dataset
    ar_id_values = ds[ar_id_var].values
    ivt_values = ds[ivt_var].values
    
    # create empty arrays
    ar_dur = np.zeros_like(ar_id_values)
    ar_max = np.zeros_like(ivt_values)
    
    # loop through ids to compute the duration for each group
    unique_ids = np.unique(ar_id_values[ar_id_values != 0])
    
    for uid in unique_ids:
        mask = ar_id_values == uid

        duration = np.sum(mask) * 6 # multiplying by 6 because we have 6-hourly data. This makes it easier to work with durations. 

        # compute max IVT for this group
        max_ivt = np.max(ivt_values[mask])
        
        # assign the duration and max IVT to all the time steps in that group
        ar_dur[mask] = duration
        ar_max[mask] = max_ivt
    
    # append the new variables to the dataset
    ds[dur_var] = (['time'], ar_dur)
    ds[max_var] = (['time'], ar_max)

    return ds
```

```python
# For table 2 of the paper
def categorize_ar_event(ds, max_var='AR_MAX', dur_var='AR_DUR', cat_var='AR_CAT'):

    # Extract the AR_MAX and AR_DUR values from the dataset
    ar_max = ds[max_var].values
    ar_dur = ds[dur_var].values
    
    # Initialize an array for AR_CAT with the same shape
    ar_cat = np.empty_like(ar_max, dtype='<U20')  # String array to store categories
    
    # Apply categorization logic based on the figure
    for i in range(len(ar_max)):
        if ar_max[i] < 250:
            ar_cat[i] = "Not an AR"
        elif ar_max[i] >= 1250:
            if ar_dur[i] <= 24:
                ar_cat[i] = "Extreme"
            elif ar_dur[i] <= 48:
                ar_cat[i] = "Exceptional"
            elif ar_dur[i] <= 72:
                ar_cat[i] = "Exceptional"
        elif ar_max[i] >= 1000:
            if ar_dur[i] <= 24:
                ar_cat[i] = "Strong"
            elif ar_dur[i] <= 48:
                ar_cat[i] = "Extreme"
            elif ar_dur[i] <= 72:
                ar_cat[i] = "Exceptional"
        elif ar_max[i] >= 750:
            if ar_dur[i] <= 24:
                ar_cat[i] = "Moderate"
            elif ar_dur[i] <= 48:
                ar_cat[i] = "Strong"
            elif ar_dur[i] <= 72:
                ar_cat[i] = "Extreme"
        elif ar_max[i] >= 500:
            if ar_dur[i] <= 24:
                ar_cat[i] = "Weak"
            elif ar_dur[i] <= 48:
                ar_cat[i] = "Moderate"
            elif ar_dur[i] <= 72:
                ar_cat[i] = "Strong"
        elif ar_max[i] >= 250:
            if ar_dur[i] <= 24:
                ar_cat[i] = "Not an AR"
            elif ar_dur[i] <= 48:
                ar_cat[i] = "Weak"
            elif ar_dur[i] <= 72:
                ar_cat[i] = "Moderate"
    
    # Add the new AR_CAT variable to the dataset
    ds[cat_var] = (['time'], ar_cat)
    
    return ds
```

```python
# sanity check
ds_subset_small_clean = categorize_ar_event(ds_subset_small_clean, max_var='AR_MAX', dur_var='AR_DUR', cat_var='AR_CAT')
ds_subset_small_clean.isel(time=50)
```

```python
import pandas as pd

ivt_values = ds_subset_small_clean['IVT'].values
ar_cat_values = ds_subset_small_clean['AR_CAT'].values
time_values = ds_subset_small_clean['time'].values

# create a dictionary to map AR_CAT string values to numbers
categories = np.unique(ar_cat_values)
ordered_categories = ["Not an AR", "Weak", "Moderate", "Strong", "Extreme", "Exceptional"]
ar_cat_values = pd.Categorical(ar_cat_values, categories=ordered_categories, ordered=True)
category_map = {cat: i for i, cat in enumerate(ordered_categories)}

# change AR_CAT to numbers for plotting on the secondary y-axis
ar_cat_numeric = np.array([category_map.get(cat, np.nan) for cat in ar_cat_values])

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(time_values, ivt_values, color='b', label='IVT')
ax1.set_xlabel('Time')
ax1.set_ylabel('IVT (kg m$^{-1}$ s$^{-1}$)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(time_values, ar_cat_numeric, marker='.', linestyle='', color='g', label='AR_CAT')
ax2.set_ylabel('AR Category', color='g')
ax2.tick_params(axis='y', labelcolor='g')
ax2.set_yticks(range(len(ordered_categories)))
ax2.set_yticklabels(ordered_categories)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```
